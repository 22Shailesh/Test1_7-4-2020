# 6. Mathematical Symbols/Latex Maths

USe this code $ $ : $Mathematical Symbols$\
Jupyter Notebooks' Markdown cells support **Latex** for formatting mathematical equations. To tell Markdown to interpret your text as **Latex**, surround yout input with *dollar signs* like this:

$z=\dfrac{2x}{3y}$

An equation can be very complex:

$F(k)=\int_{-\infty}^{\infty}f(x)e^{2\pi i k} dx$

If you want you *Latex equations to be indented towards the center of the cell*, surround your input with *two dollar signs* on each side like this:

$$2x+3y=z$$

**Mathematical signs**

1. (+) = addition
2. (-) = subtraction 
3. (/) = divide
4. (*) = multiplication
5. (**) = power
